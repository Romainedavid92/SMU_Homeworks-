```python
 # Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df.head()
```

```python
# Player Count- Finding each time a player name is listed and calling length of that list
player_count= len(purchase_data_df["SN"].unique())
player_count

# creating summary table 
summary_table = pd.DataFrame({"Total Players": [player_count]})
summary_table
```

```python
#Total number of unique items 
item_count= len(purchase_data_df["Item Name"].unique())

# Total Number of purchases
purchase_count= len(purchase_data_df["Purchase ID"].unique())

# Total Revenue
sum_purchase=purchase_data_df["Price"].sum()

# calculating average purchase price
average_purchase_price= purchase_data_df["Price"].mean()


# Creating summary table 
summary_table2 = pd.DataFrame({"Unique Item Count": [item_count],
                               "Average Purchase Price":[average_purchase_price],
                               "Total Purchases":[purchase_count],
                               "Total Revenue":[sum_purchase]
                              })
summary_table2

# Formating summary table
summary_table2["Average Purchase Price"] = summary_table2["Average Purchase Price"].map("${:,.2f}".format)
summary_table2["Total Revenue"] = summary_table2["Total Revenue"].map("${:,.2f}".format)
summary_table2
```

```python
# grouping data by gender
gender= purchase_data_df.groupby("Gender")

# Taking out duplicates
total_count_gender= gender.nunique()["SN"]

# Calculating gender percentage 
percentage_gender= total_count_gender/ player_count*100
percentage_gender

# Creating gender summary table
summary_table3=pd.DataFrame({"Total Count": total_count_gender,
                               "Percentage of Players":percentage_gender
                            })
summary_table3

# Formating summary table             
summary_table3["Percentage of Players"] = summary_table3["Percentage of Players"].map("{:,.2f}%".format)
summary_table3
   
```

```python
# grouping data by gender
gender= purchase_data_df.groupby(["Gender"]).count()["Price"]

# calculating the average purchase price by gender
average_price= purchase_data_df.groupby(["Gender"]).mean()["Price"]
average_price

# finding the sum of purchases by gender
sum_price= purchase_data_df.groupby(["Gender"]).sum()["Price"]
sum_price

# Finding the total purchase by player by gender
Total_per_person= sum_price/total_count_gender
Total_per_person


summary_table4=pd.DataFrame({"Purchase Count": gender,
                               "Average Purchase Price":average_price,
                             "Total Purchase Amount":sum_price,
                             "Total Purchase Per Player":Total_per_person
                            })
summary_table4

summary_table4["Average Purchase Price"] = summary_table4["Average Purchase Price"].map("${:,.2f}".format)
summary_table4["Total Purchase Amount"] = summary_table4["Total Purchase Amount"].map("${:,.2f}".format)
summary_table4["Total Purchase Per Player"]= summary_table4["Total Purchase Per Player"].map("{:,.2f}%".format)
summary_table4
```

```python
# Create the bins in which Data will be held
# Bins are <10,14,19,24,29,34,40+.   
bins = [0,9,14,19,24,29,34,39,100]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] 

# dropping duplicate entries
single_entry=purchase_data_df.drop_duplicates("SN", keep='first', inplace=False)
single_entry


# use the dropped entry dataset to create bins by age
single_entry["Age Groups"] = pd.cut(single_entry["Age"], bins, labels=group_names)
single_entry

# Find the number of people in each age group
age= single_entry.groupby(["Age Groups"]).count()["Age"]
age

# find the percent against the total number of entries
age_percent= (age/player_count)*100
age_percent

# another summary table! 
summary_table5=pd.DataFrame({"Total Count": age,
                               "Percentage of Players":age_percent,
                             
                            })
summary_table5


summary_table5["Percentage of Players"]= summary_table5["Percentage of Players"].map("{:,.2f}%".format)
summary_table5
```

```python
# Create the bins in which Data will be held
# Bins are <10,14,19,24,29,34,40+.   

bins = [0,9,14,19,24,29,34,39,100]

group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] 

# Creating dataframe based on age group bins
purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)


# Groupby count tells how many purhcases per age group
purchase_c= purchase_data_df.groupby(["Age Groups"]).count()["Price"]


# groupby age group take mean on price 
purchase_mean= purchase_data_df.groupby(["Age Groups"]).mean()["Price"]


# groupby age group take sum on price
purchase_sum= purchase_data_df.groupby(["Age Groups"]).sum()["Price"]


# take the purchase sum from each age group and divie by the count within each age group
avg_per_person= purchase_sum/single_entry.groupby(["Age Groups"]).count()["Price"]


# Create a summary table dataframe with above information 
summary_table6=pd.DataFrame({"Purchase Count": purchase_c,
                               "Average Purchase Price":purchase_mean,
                             "Total Purchase Value": purchase_sum,
                             "Avg Total Purchase per Person": avg_per_person
                             
                            })
summary_table6


summary_table6["Average Purchase Price"]= summary_table6["Average Purchase Price"].map("${:,.2f}".format)
summary_table6["Total Purchase Value"]= summary_table6["Total Purchase Value"].map("${:,.2f}".format)
summary_table6["Avg Total Purchase per Person"]= summary_table6["Avg Total Purchase per Person"].map("${:,.2f}".format)
summary_table6
```

```python
# Using groupby SN to obtain purchase count
new= purchase_data_df.groupby(["SN"]).count()["Price"]


# using groupby to calculate mean of price for each SN
mean= purchase_data_df.groupby(["SN"]).mean()["Price"]


# using grouby to find total purchase value for each SN
sum = purchase_data_df.groupby(["SN"]).sum()["Price"]

summary_table7=pd.DataFrame({"Purchase Count": new,
                               "Average Purchase Price":mean,
                             "Total Purchase Value": sum
                            }
                           )
summary_table7

# Arranging summary table 7 to have highest total purchase value at top
summary_table7 = summary_table7.sort_values("Total Purchase Value", ascending=False)
summary_table7


summary_table7["Average Purchase Price"]= summary_table7["Average Purchase Price"].map("${:,.2f}".format)
summary_table7["Total Purchase Value"]= summary_table7["Total Purchase Value"].map("${:,.2f}".format)
summary_table7.head()
```

```python
# Using grouby to obtain item id and item name count 
item_count= purchase_data_df.groupby(["Item ID","Item Name"]).count()["Price"]


# using groupby to organize by id and item name and then finding the sum of price per item
item_purchase_value= purchase_data_df.groupby(["Item ID","Item Name"]).sum()["Price"]



summary_table8=pd.DataFrame({"Purchase Count": item_count,
                               "Total Purchase Value": item_purchase_value,
                            }
                           )
summary_table8


# using the merge function to merge the orginal table and summary table 8 together on Item ID
merge_table = pd.merge(purchase_data_df, summary_table8, on="Item ID")


# Calling only the columns I want from the newly merged table
merge_table1= merge_table[["Item ID","Item Name","Purchase Count","Price","Total Purchase Value"]]


# dropping duplicate entries
single_item=merge_table1.drop_duplicates("Item ID", keep='first', inplace=False)
single_item


# Set new index to Item ID
ID_index = single_item.set_index("Item ID")


# arranging the merged table to have the highest purchase counts at top
summary_table9 = ID_index.sort_values("Purchase Count", ascending=False)
summary_table9


summary_table9["Price"]= summary_table9["Price"].map("${:,.2f}".format)
summary_table9["Total Purchase Value"]= summary_table9["Total Purchase Value"].map("${:,.2f}".format)
summary_table9.head()
```
